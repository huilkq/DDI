```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
sns.set(style='ticks')
sns.set(color_codes=True)
sns.set_style("white")
```

```python
predictresult = pd.read_csv("GMDZT_SM/_GMDZT_SM_predict_task3_DDIMDL.csv")
predictresult
```

```python
predictresult['Label_1'].sum()  # D_1列'1'的总和
predictresult.shape[0]-predictresult['Label_1'].sum() # D_1列'0'的总
```

```python

plt.figure(figsize=(8,6)) #调节图形大小
#labels = ['metabolism increase','metabolism decrease'] #定义标签 
#labels = ['serum concentration decrease','serum concentration increase']
labels = ['therapeutic efficacy increase','therapeutic efficacy decrease']
sizes = [predictresult['Label_1'].sum(),predictresult.shape[0]-predictresult['Label_1'].sum()] #每块值
colors = ['#f1ccb8','#f1b8e4'] #每块颜色定义
explode = (0,0.05) #将某一块分割出来，值越大分割出的间隙越大
patches,l_text,p_text = plt.pie(sizes,
                      explode=explode,
                      labels=labels,
                      colors=colors,
                      autopct = '%.2f%%', #数值保留固定小数位
                      shadow = False, #无阴影设置
                      startangle = 90, #逆时针起始角度设置
                      pctdistance = 0.6) #数值距圆心半径倍数的距离
#patches饼图的返回值，texts1饼图外label的文本，texts2饼图内部的文本
# x，y轴刻度设置一致，保证饼图为圆形

#改变文本的大小
#方法是把每一个text遍历。调用set_size方法设置它的属性
for t in l_text:
    t.set_size(20) # 标签BIG等
for t in p_text:
    t.set_size(15) # 6.71%

plt.axis('equal') #  # 显示为圆（避免比例压缩为椭圆）
plt.savefig("task_3.png", facecolor ="w", dpi=300, bbox_inches = 'tight')
# 设置背景facecolor为白色; plt.savefig要放在show()前面，不然放在后面为全黑色; bbox_inches设为tight防止标签被截断
plt.show()
```
